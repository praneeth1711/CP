```python
def subset_sum_mitm(arr, target):
    n = len(arr)
    mid = n // 2

    left = arr[:mid]
    right = arr[mid:]

    left_sums = set()
    for mask in range(1 << len(left)):
        s = 0
        for i in range(len(left)):
            if mask & (1 << i):
                s += left[i]
        left_sums.add(s)

    for mask in range(1 << len(right)):
        s = 0
        for i in range(len(right)):
            if mask & (1 << i):
                s += right[i]
        if target - s in left_sums:
            return True

    return False


t = int(input())

for _ in range(t):
    n, s = map(int, input().split())
    arr = list(map(int, input().split()))

    print("YES" if subset_sum_mitm(arr, s) else "NO")
```
